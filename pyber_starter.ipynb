```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data=pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber=pd.merge(ride_data,city_data,how='left',on=['city'])
# Display the data table for preview
pyber.head()
```

```python
urban_data=pyber.loc[pyber['type']=='Urban']
suburban_data=pyber.loc[pyber['type']=='Suburban']
rural_data=pyber.loc[pyber['type']=='Rural']
```

## Bubble Plot of Ride Sharing Data

```python
fare_avg_urban=urban_data.groupby('city')['fare'].mean()
fare_avg_suburban=suburban_data.groupby('city')['fare'].mean()
fare_avg_rural=rural_data.groupby('city')['fare'].mean()
```

```python
total_rides_urban=urban_data.groupby('city').count()['ride_id']
total_rides_suburban=suburban_data.groupby('city').count()['ride_id']
total_rides_rural=rural_data.groupby('city').count()['ride_id']
```

```python
total_drivers_urban=urban_data.groupby('city')['driver_count'].value_counts()
total_drivers_suburban=suburban_data.groupby('city')['driver_count'].value_counts()
total_drivers_rural=rural_data.groupby('city')['driver_count'].value_counts()
```

```python
plt.scatter(fare_avg_urban,total_rides_urban,s=total_drivers_urban*10,
            marker='o',facecolors="red",edgecolors='black',alpha=0.5,label="Urban")

plt.scatter(fare_avg_suburban,total_rides_suburban,s=total_drivers_suburban*10,
            marker='o',facecolors="green",edgecolors='black',alpha=0.5,label="Suburban")

plt.scatter(fare_avg_rural,total_rides_rural,s=total_drivers_rural*10,
            marker='o',facecolors='lightblue',edgecolors='black',alpha=0.5,label="Rural")

plt.title("Pyber")
plt.xlabel("Total Rides per Driver")
plt.ylabel("Average Fare $")
plt.legend(loc='best')

plt.savefig("Pyber Ahmad.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_faretype=pyber.groupby('type')['fare'].sum()
labels=['Rural','Suburban','Urban']
colors=['lightblue','green','red']
explode=(0,0,0.1)

# Build Pie Chart
plt.pie(total_faretype,labels=labels,explode=explode,colors=colors,shadow=True,autopct='%1.1f%%',startangle=140)
plt.title("% of Total Fares by City Type")
plt.axis('equal')

# Save Figure
plt.savefig('% of Total Fares by City Type.png')
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_ridetype=pyber.groupby('type')['ride_id'].count()
labels=['Rural','Suburban','Urban']
colors=['lightblue','green','red']
explode=(0,0,0.1)

# Build Pie Chart
plt.pie(total_ridetype,labels=labels,explode=explode,colors=colors,shadow=True,autopct='%1.1f%%',startangle=140)
plt.title("% of Total Rides by City Type")
plt.axis('equal')
# Save Figure
plt.savefig('% of Total Rides by City Type.png')
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
total_drivertype=city_data.groupby('type')['driver_count'].sum()
labels=['Rural','Suburban','Urban']
colors=['lightblue','green','red']
explode=(0,0,0.1)
# Build Pie Charts
plt.pie(total_drivertype,labels=labels,explode=explode,colors=colors,shadow=True,autopct='%1.1f%%',startangle=140)
plt.title("% of Total Drivers by City Type")
plt.axis('equal')
# Save Figure
plt.savefig('% of Total Drivers by City Type.png')
plt.show()
```
